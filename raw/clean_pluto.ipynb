```python
from IPython.core.interactiveshell import InteractiveShell
import numpy as np
import pandas as pd
import seaborn as sns

InteractiveShell.ast_node_interactivity = "all"
```

```python
%%time

! cd data; ls

plutos = {}
for boro in ['BK', 'BX', 'MN', 'QN', 'SI']:
    print(f'\n Loading {boro}...')
    plutos[boro] = pd.read_csv(f'pluto{boro}.csv', skiprows=None, nrows=None, low_memory=False)

print('\n Concatenating...')
pluto = pd.concat(list(plutos.values()), sort=False)
pluto = pluto.reset_index(drop=True)

print('\n Pickling... \n')
pluto.to_pickle('../pluto_full.pkl')
```
